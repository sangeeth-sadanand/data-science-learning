# Creating a Simple Line Plot

## Import Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Prepare data

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

## Plot the data

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.title("My First Line Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.show()
```

## **1.BasicCustomizationsin**`plot()`

Youcanspecify**color**,**linestyle**,and**marker**directlyusingformatstringsorkeywordarguments

### **ShorthandStyleString**

```python
plt.plot(x, y, 'r--')  # red dashed line
```

### **KeywordArguments**

```python
plt.plot(x, y, color='purple', linestyle=':', marker='o')
```

### **2.CommonOptions**

| Property     | ExampleValues                                                  |
| ------------ | --------------------------------------------------------------- |
| `color`      | `'blue'`,`'r'`,`'#1f77b4'`                                    |
| `linestyle`  | `'-'`(solid),`'--'`(dash),`':'`(dotted),`'-.'`(dash-dot) |
| `linewidth`  | `1`,`2.5`,etc.                                                |
| `marker`     | `'o'`(circle),`'^'`(triangle),`'s'`(square),`'*'`,`'x'`  |
| `markersize` | `5`,`10`,etc.                                                 |

#### Example:CustomLineStyle


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, color='orange', linestyle='-.', linewidth=2, marker='s', markersize=8)
plt.title("Customized Line Plot")
plt.grid(True)
plt.show()
```

#### MultipleLineswithDifferentStyles

```python
plt.plot(x, y, 'b-o', label="Data 1")
plt.plot(x, [i*2 for i in y], 'g--^', label="Data 2")
plt.legend()
```
