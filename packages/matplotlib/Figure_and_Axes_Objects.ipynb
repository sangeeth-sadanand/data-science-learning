# ðŸ§± FigureÂ andÂ Axes:Â TheÂ CoreÂ BuildingÂ Blocks

MatplotlibÂ plotsÂ areÂ builtÂ onÂ twoÂ essentialÂ objectÂ types:

## 1.Â **FigureÂ (**`matplotlib.figure.Figure`**)**
- TheÂ **entireÂ canvas**Â onÂ whichÂ everythingÂ isÂ drawn.
- ThinkÂ ofÂ itÂ likeÂ aÂ blankÂ pageÂ orÂ dashboard.
- CanÂ containÂ oneÂ orÂ moreÂ **Axes**

### 2.Â **AxesÂ (**`matplotlib.axes.Axes`**)**
- TheÂ actualÂ **plotÂ orÂ graph**Â whereÂ dataÂ isÂ visualized.
- AÂ singleÂ figureÂ canÂ haveÂ **multipleÂ axes**Â (e.g.,Â subplots).
- EachÂ AxesÂ objectÂ hasÂ itsÂ ownÂ x-Â andÂ y-axis.

> [!info]
Don'tÂ confuseÂ AxesÂ withÂ axisÂ â€”Â **Axes**Â isÂ theÂ plotÂ area,Â **axis**Â refersÂ toÂ x/yÂ lines.


## Â **CreatingÂ aÂ FigureÂ andÂ Axes**

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()  # returns a Figure and a single Axes
ax.plot([1, 2, 3], [1, 4, 9])
ax.set_title("Using Figure and Axes")
plt.show()
````



 ```fig`:Â theÂ FigureÂ object
- `ax`:Â theÂ AxesÂ objectÂ (orÂ anÂ arrayÂ ofÂ themÂ ifÂ you'reÂ creatingÂ subplots)

## ðŸ§©Â **MultipleÂ Axes

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot([1, 2, 3], [1, 2, 3])
ax1.set_title("Line 1")

ax2.plot([1, 2, 3], [1, 4, 9])
ax2.set_title("Line 2")

plt.tight_layout()
plt.show()
````

- `subplots(rows,Â cols)`Â createsÂ multipleÂ AxesÂ inÂ oneÂ Figure  
- `tight_layout()`Â adjustsÂ spacingÂ toÂ preventÂ overlap

## Subplots using `plt.subplot()` and `plt.subplots()

### `plt.subplot()`Â â€“Â **ManualÂ subplotÂ positioning**

```python
plt.subplot(nrows, ncols, index)
```

- CreatesÂ theÂ `index`-thÂ subplotÂ inÂ aÂ gridÂ ofÂ `nrowsÂ Ã—Â ncols`.

````python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.plot([1, 2], [3, 4])
plt.title("Plot 1")

plt.subplot(1, 2, 2)  # 2nd subplot
plt.plot([1, 2], [4, 3])
plt.title("Plot 2")

plt.tight_layout()
plt.show()
````

### `plt.subplots()`Â â€“Â **TheÂ modern,Â flexibleÂ way**
```python
fig, ax = plt.subplots(nrows, ncols)
```

#### **AdvantagesÂ of**Â `subplots()`
- YouÂ getÂ fullÂ accessÂ toÂ theÂ `Figure`Â andÂ `Axes`Â objects
- BetterÂ layoutÂ controlÂ (e.g.,Â `figsize`,Â `gridspec`,Â sharedÂ axes)
- ScalesÂ wellÂ toÂ largerÂ plotsÂ andÂ apps

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(8, 6))  # 2x2 grid
axs[0, 0].plot([1, 2], [1, 4])
axs[0, 0].set_title("Top Left")

axs[1, 1].plot([1, 2], [2, 1])
axs[1, 1].set_title("Bottom Right")

fig.suptitle("Multiple Subplots Example")
plt.tight_layout()
plt.show()
````

## Layout Adjustments (`tight_layout`, `gridspec`)
### `tight_layout()`Â â€“Â QuickÂ Auto-FixÂ forÂ OverlappingÂ Elements

ThisÂ isÂ yourÂ go-toÂ commandÂ forÂ automaticallyÂ adjustingÂ spacingÂ betweenÂ subplotsÂ toÂ preventÂ overlapÂ ofÂ titles,Â labels,Â andÂ ticks.

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2)
fig.suptitle("My Grid of Plots")

plt.tight_layout()  # Automatically adjusts spacing
plt.show()
````

````python


import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2)
fig.suptitle("My Grid of Plots")

plt.tight_layout()
fig.subplots_adjust(top=0.88)
plt.show()
````

### ðŸ§°Â `GridSpec`Â â€“Â Fine-GrainedÂ LayoutÂ Control

ForÂ moreÂ advancedÂ gridÂ customizationsâ€”likeÂ makingÂ oneÂ subplotÂ spanÂ multipleÂ rows/columnsâ€”youâ€™llÂ wantÂ `matplotlib.gridspec`.

- PreciseÂ controlÂ overÂ subplotÂ sizesÂ andÂ placements
- IdealÂ whenÂ youÂ needÂ asymmetricalÂ gridsÂ (e.g.,Â dashboardÂ layouts)
- WorksÂ wellÂ withÂ manuallyÂ resizedÂ figures

````python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(3, 3)

ax1 = fig.add_subplot(gs[0, :])     # Top row spans all columns
ax2 = fig.add_subplot(gs[1:, 0])    # Left column spans two rows
ax3 = fig.add_subplot(gs[1:, 1:])   # Bottom-right 2x2

ax1.set_title("Top Wide Plot")
ax2.set_title("Tall Side Plot")
ax3.set_title("Main Plot")

plt.tight_layout()
plt.show()
````
